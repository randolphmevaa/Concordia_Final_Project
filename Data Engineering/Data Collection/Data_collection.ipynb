```python
Jurisdiction = []
for i in range(len(df['Address'])):
    Jurisdiction.append(','.join(df['Address'].str.split(',')[i][-2:]))

df['Jurisdiction'] = Jurisdiction
```

```python
Jurisdiction2 = []
for i in range(len(df['Jurisdiction'])):
    Jurisdiction2.append(df['Jurisdiction'][i][1:-6])

df['Jurisdiction'] = Jurisdiction2
```

```python
df
```

```python
Zip_Code = []
for i in range(len(df['Address'])):
    Zip_Code.append(df['Address'].str.split(' ')[i][-1])

df['Zip Code'] = Zip_Code
```

```python
State = []
for i in range(len(df['Address'])):
    State.append('California')

df['State'] = State
```

```python
import numpy as np
from geopy.exc import GeocoderTimedOut
from geopy.geocoders import Nominatim

def findGeocode(city):

    try:
        geolocator = Nominatim(user_agent="your_app_name")
        return geolocator.geocode(city)
    
    except GeocoderTimedOut:
        return findGeocode(city)    

```

```python
longitude = []
latitude = []

for i in (df['Jurisdiction']):
      
    if findGeocode(i) != None:
        loc = findGeocode(i)
        latitude.append(loc.latitude)
        longitude.append(loc.longitude)
    else:
        latitude.append(np.nan)
        longitude.append(np.nan)
```

```python
df['Longitude'] = longitude
df['Latitude'] = latitude
df
```
