```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://www.laalmanac.com/crime/cr03.php'

r = requests.get(url)
soup = bs(r.content, 'html.parser')
```

```python
#Table head

thead = soup.find_all('thead')[1]
head_title = []

for i in range(len(thead.find_all('th'))):
    head_title.append(thead.find_all('th')[i].text)

# We won't use this table head because of the Measure columns, we only want Crime rates
```

```python
# Jurisdiction
Jurisdiction = []

for i in range(len(soup.find_all('tbody'))):
    tbody = soup.find_all('tbody')[i]
    Jurisdiction.append(tbody.find_all('td', class_="text-left")[0].text+', CA')
    Jurisdiction.append(tbody.find_all('td', class_="text-left")[3].text+', CA')
    Jurisdiction.append(tbody.find_all('td', class_="text-left")[6].text+', CA')

# Jurisdiction
```

```python
# Total Violent Crimes 
Total_Violent_Crimes = []

for i in range(len(soup.find_all('tbody'))):
    tbody = soup.find_all('tbody')[i]
    tr = tbody.find_all('tr')[1]
    tr1 = tbody.find_all('tr')[3]
    tr2 = tbody.find_all('tr')[5]
    Total_Violent_Crimes.append(tr.select('td')[1].text)
    Total_Violent_Crimes.append(tr1.select('td')[1].text)
    Total_Violent_Crimes.append(tr2.select('td')[1].text)

# Total_Violent_Crimes
```

```python
# Homicide
Homicide = []

for i in range(len(soup.find_all('tbody'))):
    tbody = soup.find_all('tbody')[i]
    tr = tbody.find_all('tr')[1]
    tr1 = tbody.find_all('tr')[3]
    tr2 = tbody.find_all('tr')[5]
    Homicide.append(tr.select('td')[2].text)
    Homicide.append(tr1.select('td')[2].text)
    Homicide.append(tr2.select('td')[2].text)

# Homicide
```

```python
# Rape (Forcible Rape prior to 2014)
Rape = []

for i in range(len(soup.find_all('tbody'))):
    tbody = soup.find_all('tbody')[i]
    tr = tbody.find_all('tr')[1]
    tr1 = tbody.find_all('tr')[3]
    tr2 = tbody.find_all('tr')[5]
    Rape.append(tr.select('td')[3].text)
    Rape.append(tr1.select('td')[3].text)
    Rape.append(tr2.select('td')[3].text)

# Rape
```

```python
# Robbery
Robbery = []

for i in range(len(soup.find_all('tbody'))):
    tbody = soup.find_all('tbody')[i]
    tr = tbody.find_all('tr')[1]
    tr1 = tbody.find_all('tr')[3]
    tr2 = tbody.find_all('tr')[5]
    Robbery.append(tr.select('td')[4].text)
    Robbery.append(tr1.select('td')[4].text)
    Robbery.append(tr2.select('td')[4].text)

# Robbery
```

```python
# Aggravated Assault
Assault = []

for i in range(len(soup.find_all('tbody'))):
    tbody = soup.find_all('tbody')[i]
    tr = tbody.find_all('tr')[1]
    tr1 = tbody.find_all('tr')[3]
    tr2 = tbody.find_all('tr')[5]
    Assault.append(tr.select('td')[5].text)
    Assault.append(tr1.select('td')[5].text)
    Assault.append(tr2.select('td')[5].text)

# Assault
```

```python
data = {
    'Jurisdiction': Jurisdiction, 
    'Total Violent Crimes': Total_Violent_Crimes,
    'Homicide' : Homicide, 
    'Rape (Forcible Rape prior to 2014)': Rape, 
    'Robbery': Robbery,
    'Aggravated Assault': Assault
    }
# data
```

```python
# Build the df and export it into a csv
df2 = pd.DataFrame(data)
df2.Jurisdiction[0] = 'County Total, CA'
df2
```

```python
import numpy as np
from geopy.exc import GeocoderTimedOut
from geopy.geocoders import Nominatim

def findGeocode(city):

    try:
        geolocator = Nominatim(user_agent="your_app_name")
        return geolocator.geocode(city)
    
    except GeocoderTimedOut:
        return findGeocode(city)    
```

```python
longitude = []
latitude = []

for i in (df2['Jurisdiction']):
      
    if findGeocode(i) != None:
        loc = findGeocode(i)
        latitude.append(loc.latitude)
        longitude.append(loc.longitude)
    else:
        latitude.append(np.nan)
        longitude.append(np.nan)
```

```python
df2['Longitude'] = longitude
df2['Latitude'] = latitude
```

```python
df2 = df2.dropna()
df2.to_csv('Crimes_rates')
```

```python
url2 = 'https://www.laalmanac.com/population/po24la.php'

html = requests.get(url2)
webpage = bs(html.content, 'html.parser')
```

```python
# Table head name

head = webpage.find_all('thead')[0]
head_name = head.find('th', class_="text-left").text
```

```python
# Neighborhood
Neighborhood = []

for i in range(49):
    body = webpage.find_all('tbody')[i]
    tr0 = body.select('tr', class_="text-left")[0]
    tr2 = body.select('tr', class_="text-left")[2]
    tr4 = body.select('tr', class_="text-left")[4]
    Neighborhood.append(tr0.find('td').text+', Los Angeles')
    Neighborhood.append(tr2.find('td').text+', Los Angeles')
    Neighborhood.append(tr4.find('td').text+', Los Angeles')

# Neighborhood
```

```python
for i in range(49, 50):
    body = webpage.find_all('tbody')[i]
    tr0 = body.select('tr', class_="text-left")[0]
    tr2 = body.select('tr', class_="text-left")[2]
    Neighborhood.append(tr0.find('td').text+', CA')
    Neighborhood.append(tr2.find('td').text+', CA')

len(Neighborhood)
```

```python
# Table head name 2

head2 = webpage.find_all('thead')[0]
head_name_2 = head2.select('th', attrs={'rowspan':'2'} )[1].text
```

```python
# Total Population, All Races
Total_population = []

for i in range(49):
    body = webpage.find_all('tbody')[i]
    tr0 = body.find_all('tr')[0]
    tr2 = body.find_all('tr')[2]
    Total_population.append(tr0.select('td')[1].text)
    Total_population.append(tr2.select('td')[1].text)
    Total_population.append(body.find_all('tr')[4].select('td')[1].text)

len(Total_population)
```

```python
for i in range(49,50):
    body = webpage.find_all('tbody')[i]
    tr0 = body.find_all('tr')[0]
    tr2 = body.find_all('tr')[2]
    Total_population.append(tr0.select('td')[1].text)
    Total_population.append(tr2.select('td')[1].text)

len(Total_population)
```
